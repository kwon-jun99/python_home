```python
missing_count=0                             #성적 None(비정상적 데이터) 정의

for student in students:                    #None 갯수 파악후 확인/들여쓰기 잘되어 있는지 확인
    if student["성적"] is None:
        missing_count+=1

print(f"결측치 갯수 {missing_count}이다")     #print로 결과 확인/ 원래코드 print("결측치 갯수", missing_count, "이다")  
```

```python
scores = []                                 #성적 정의(평균 계산을 위해 유효한 성적만 저장할 빈 리스트 만들기)

for student in students:                    #정상 데이터의 합 구하기
    if student["성적"] is not None:
        scores.append(student["성적"])

avg = round(sum(scores) / len(scores))      #평균 구하기(점수는 정수여야하기 때문에 round로 반올림)/원래코드 avg = sum(scores) / len(scores)

for student in students:
     if student["성적"] is None:
         student["성적"] = avg               #jupyter notebook은 메모리가 저장되기 때문에 한번 저장된 데이터가 안 지워짐
                                            #그래서 중간에 수정 후 실행하고 싶으면 상단에 restart로 커널 리셋하고 run all로 처음부터 다시 실행하면 됨.(py파일은 상관없음)

students
```
